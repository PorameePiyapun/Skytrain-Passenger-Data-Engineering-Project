```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set(style="whitegrid")
```

```python
file_path = "D:\\coding\\data_en\\data\\processed\\Skytrain_data_Processed_20250121.csv"  # Replace with your file path
data = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=data, x='passengers_entering', hue='station', fill=True, common_norm=False)
plt.title('Density of Passengers Entering by Station', fontsize=16)
plt.xlabel('Number of Passengers Entering', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Station')
plt.show()
```

```python
data['Datetime'] = pd.to_datetime(data['Date'] + ' ' + data['time'])
data['Hour'] = data['Datetime'].dt.hour

def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

data['Daytime'] = data['Hour'].apply(categorize_time)
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=data, x='Hour', hue='Daytime', fill=True, common_norm=False)
plt.title('Density of Passenger Traffic by Time of Day', fontsize=16)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Daytime')
plt.show()
```
